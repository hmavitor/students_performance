```python
# codificando binárias
mapping = {'yes':1,'no':0, 'GP':1,'MS':0, 'M':1, 'F':0, 'R':1,'U':0}
df_por_01 = df_por.replace(mapping)
df_por_01.tail()
```

```python
# criando coluna de média para desempenho acadêmico
df_por_01['Gm'] = df_por_01[['G1','G2','G3']].mean(axis=1)
df_por_01.head()
```

```python
# visualização para detectar possível associação de Dalc com nominais
df_por_01.groupby('Dalc')['school'].value_counts(normalize=True).plot.bar()
```

Foi feita uma análise visual se o indicador de consumo alcoólico (Dalc) apresentava diferença entre variáveis nominais (binárias ou não) que não foram codificadas e não poderiam entrear na correlação. Exemplos: Pstatus, guardian, Medu, Fedu, Mjob, Fjob, famsize, school. Não foi verificado nenhum indicativo de associação. Portanto, essas variáveis não foram codificadas para testes de associação estatísticos.

```python
df_por_num = df_por_01.select_dtypes(include=['float64', 'int64'])
df_por_num.head()
```

```python
df_por_corr = df_por_num.corr(method='spearman')
```

```python
plt.figure(figsize=(30, 25))
sns.heatmap(df_por_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Mapa de Calor de Correlação (Spearman)')
plt.show()
```
